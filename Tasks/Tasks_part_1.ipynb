### Task 7: Using Having clause, find out the total sharewomen greater than 5

```python
connection = sql.connect('jobs.db')

connection
```

```python
query = 'SELECT * FROM sqlite_master'

pd.read_sql_query(query, connection)
```

```python
query = """
SELECT SUM(ShareWomen) AS 'Total_ShareWomen', Major_category
FROM recent_grads
GROUP BY Major_category
HAVING Total_ShareWomen > 5
"""
pd.read_sql_query(query, connection)
```

### Task 8: Create a column with a men's percentage.

```python
query = """
SELECT ShareWomen, Men, Women, Total,
round((cast(Men as float)/Total)*100, 3) AS "Men's %"
FROM recent_grads
"""

pd.read_sql_query(query, connection)
```

### Task 9: Create a column having population percentage of cities table from factbook.db.

```python
connection = sql.connect('factbook.db')

connection
```

```python
query = """
SELECT * FROM cities
"""

pd.read_sql_query(query, connection)
```

```python
query = """
SELECT name, population,
(cast(population as float) / (SELECT SUM(population) FROM cities))*100 AS 'Population_%'
FROM cities
"""

pd.read_sql_query(query, connection)
```
