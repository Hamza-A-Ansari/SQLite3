### Task 11: Top 10 countries having maximum population according to the cities table.

```python
query = """
SELECT id, name, population, facts_id
FROM cities
"""

pd.read_sql_query(query, connection)
```

```python
query = """
SELECT facts_id AS Country_id, SUM(population) AS Total_Pop
FROM cities
GROUP BY facts_id
ORDER BY SUM(population) DESC
LIMIT 10
"""

pd.read_sql_query(query, connection)
```

### Task 12: Create a result set having the following columns: Country Names, Country Population, Urban Population and Population Percentage.

```python
query = """
SELECT 
    facts.name AS 'Country Names', 
    facts.population AS 'Country Population',
    SUM(cities.population) AS 'Urban Population',
    round((SUM(cities.population) / CAST(facts.population AS FLOAT))*100, 2) AS 'Population Percentage'
    
FROM facts

JOIN cities
ON facts.id = cities.facts_id

GROUP BY facts.name
HAVING facts.population > (SUM(cities.population))
ORDER BY round((SUM(cities.population) / CAST(facts.population AS FLOAT))*100, 2) DESC
"""

pd.read_sql_query(query, connection)
```

### Class Task: Add Other percentage Column

```python
query = """
SELECT 
    facts.name AS 'Country Names', 
    facts.population AS 'Country Population',
    SUM(cities.population) AS 'Urban Population',
    round((SUM(cities.population) / CAST(facts.population AS FLOAT))*100, 2) AS 'Population Percentage',
    100-(round((SUM(cities.population) / CAST(facts.population AS FLOAT))*100, 2)) AS 'Other Percentage'
    
FROM facts

JOIN cities
ON facts.id = cities.facts_id

GROUP BY facts.name
HAVING facts.population > (SUM(cities.population))
ORDER BY round((SUM(cities.population) / CAST(facts.population AS FLOAT))*100, 2) DESC
"""

pd.read_sql_query(query, connection)
```
