```python
# Employee_Full_Name
# Employee_Title
# Customer_Full_Name
# Invoice_Id
# Track_Name
# Artist_Name
# Quantity
# Price
q = """
SELECT 
     employee.first_name || " " || employee.last_name Employee_Full_Name,
     employee.title Employee_Title,
     customer.first_name || " " || customer.last_name Customer_Full_Name,
     invoice.invoice_id Invoice_Id,
     track.name Track_Name,
     artist.name Artist_Name,
     invoice_line.Quantity Quantity,
     invoice_line.unit_price Price
     
FROM invoice

INNER JOIN customer
ON invoice.customer_id = customer.customer_id

INNER JOIN employee
ON employee.employee_id = customer.support_rep_id

INNER JOIN invoice_line
ON invoice_line.invoice_id = invoice.invoice_id

INNER JOIN track
ON invoice_line.track_id = track.track_id

INNER JOIN album
ON album.album_id = track.album_id

INNER JOIN artist
ON artist.artist_id = album.artist_id

WHERE invoice.invoice_id == 1

"""

pd.read_sql_query(q, connection)
```

```python
invoice_id_1.to_csv('invoice_id_1.csv', index=False)
```

```python
# Identify the most selling track amoung all
# Track Name
# Track ID
# Unit Price

q = """
SELECT 
     track.name Track_Name,
     track.track_id,
     invoice_line.unit_price Price
     
FROM track

INNER JOIN invoice_line
ON track.track_id = invoice_line.track_id

GROUP BY track.track_id

ORDER BY SUM(invoice_line.quantity) DESC
LIMIT 1

"""

most_selling_track = pd.read_sql_query(q, connection)
most_selling_track
```

```python
most_selling_track.to_csv('most_selling_track.csv', index=False)
```

```python
# Identify the Country to be consider for discount/offer.
# Country name
# Cities in the country
q = """
SELECT
    customer.country
FROM customer

INNER JOIN invoice
ON invoice.customer_id == customer.customer_id

GROUP BY COUNTRY
ORDER BY SUM(invoice.total)

"""
pd.read_sql_query(q, connection)
```
